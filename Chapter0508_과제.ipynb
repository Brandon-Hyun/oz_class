## 모험가들의 레벨업 도전

### [배경 이야기]

당신은 마법사, 전사, 도적, 힐러 네 명의 모험가를 이끄는 리더입니다.

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

<br>

### [동료 캐릭터와 능력치 정보]

마법사: 힘(10), 민첩(20), 지능(80), 체력(50)

전사: 힘(80), 민첩(30), 지능(20), 체력(100)

도적: 힘(40), 민첩(80), 지능(30), 체력(60)

힐러: 힘(20), 민첩(40), 지능(70), 체력(70)

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
Wizard_stats_Str, Wizard_stats_Dex, Wizard_stats_Int, Wizard_stats_Hp = 10, 20, 80, 50
Warrior_stats_Str, Warrior_stats_dex, Warrior_stats_Int, Warrior_stats_Hp = 80, 30, 20, 100
Thief_stats_Str, Thief_stats_Dex, Thief_stats_Int, Thief_stats_Hp = 40, 80, 30, 60
Healer_stats_Str, Healer_stats_Dex, Healer_stats_Int, Healer_stats_Hp = 20,40, 70, 70

if Wizard_stats_Int >= 50:
    print(F'지능이 50 이상입니다.  현재 지능은 {Wizard_stats_Int} 입니다.')
else: print(f'마법사의 지능이 50 이상이 아닙니다.\n현재 지능은 {Wizard_stats_Int} 입니다.')
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
if Warrior_stats_Str >= 60 and Warrior_stats_Hp >= 80:
    print(f'전사의 힘이 60 이상, 체력이 80 이상입니다.')
else: print(f'전사의 힘이 60 이상, 체력이 80 이상이 아닙니다.\n현재 힘은 {Warrior_stats_Str}, 체력은 {Warrior_stats_Hp}입니다')
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
if Thief_stats_Dex >= 70 or Thief_stats_Str >= 50:
    print(f'도적의 민첩이 70 이상이거나 힘이 50 이상입니다.')
else: print(f'도적의 민첩이 70 이상이거나 힘이 50 이상이 아닙니다.\n현재 민첩은 {Thief_stats_Dex}. 힘은 {Thief_stats_Str}입니다.')
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
if Healer_stats_Int >= 60 and Healer_stats_Hp >= 50:
    print(f'힐러의 지능이 60 이상이고, 체력이 50 이상입니다.\n현재 지능은 {Healer_stats_Int}, 체력은 {Healer_stats_Hp}입니다.')
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
# 변수 선언
Wizard_stats_Exp, Warrior_stats_Exp, Thief_stats_Exp, Healer_stats_Exp = 120, 150, 110, 90
# 마법사
if Wizard_stats_Exp >= 100 :
    print(f'현재 마법사의 경험치: {Wizard_stats_Exp}\n마법사의 경험치가 100 이상으로 레벨업 합니다')

# Warrior
if Warrior_stats_Exp >= 100 :
    print(f'현재 전사의 경험치: {Warrior_stats_Exp}\n전사의 경험치가 100 이상으로 레벨업 합니다')

# 힐러
if Thief_stats_Exp >= 100 :
    print(f'현재 도적의 경험치: {Thief_stats_Exp}\n도적의 경험치가 100 이상으로 레벨업 합니다')

# 힐러
if Healer_stats_Exp >= 100 :
    print(f'현재 힐러의 경험치: {Healer_stats_Exp}\n힐러의 경험치가 100 이상으로 레벨업 합니다')
```
