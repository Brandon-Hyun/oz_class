```python
#2. if문 문법과 주의하상 (콜론, 들여쓰기)
개수 = int(input("붕어빵 몇 개 사실건가요?"))

if 개수 > 3:
        print("개당 1800원 입니다.")
```

```python
#3. if문의 흐름 (작동원리)
# 코드는 위에서 아래 순서대로 작동합니다.

# 2)개수라는 변수에 담음      1) 입력을 받음
개수 = 개수 = int(input("붕어빵 몇 개 사실건가요?"))
# 3) 개수 변수가 있네 값이 뭐지   4>4  True 또는 False

if 개수 > 3:
    print("개당 1800원 입니다")

if 개수 <= 3:
    print("개당 2000원 입니다.")
```

```python
#5. if ~ else 문법과 주의사항

개수 = 개수 = int(input("붕어빵 몇 개 사실건가요?"))

if 개수 > 3:
    print("개당 1800원 입니다")
else:
    print("개당 2000원 입니다.")
```

```python
#6. if ~ else문 흐흠(작동원리)

if 개수 > 3:
    print("개당 1800원 입니다")
else:
    print("개당 2000원 입니다.")
```

```python
# 7. 조건문의 조건 (True, False)

#True
if True:
    print(True)
else:
    print(False)

#True
if True:
    print(True)
else:
    print(False)   #>>> false

#none
if True:
    print(True)
else:
    print(False)   #>>> False

# 0
if True:
    print(True)
else:
    print(False) #>>> False

# 1
if True:
    print(True)
else:
    print(False) #>>> True

# 1.1
if True:
    print(True)
else:
    print(False) #>>>True

# Hello
if True:
    print(True)
else:
    print(False) #>>> True  모든 문자열은 True

# ' '
if True:
    print(True)
else:
    print(False)  #>>> False 빈 값이면 False

# not
if not True:   #>>> True -> False
    print(True)
else:
    print(False)

if not ' ':
    print(True)
else:
    print(False)
```

```python
#9. elif ~ if문

맛 = input("붕어빵 맛을 골라주세요(팥, 슈크림, 잡채)")
if 맛 == "팥":
    print(f'{맛}을 선택하셨습니다.')
elif 맛 == "슈크림":
    print(f'{맛}을 선택하셨습니다.')
else:
    print(f'{맛}을 선택하셨습니다.')
```

```python
#10 예제로 if문 익혀보기
# 문제 : 단어 a의 가운데 글자를 반환하느 코드
#조건 : 단어의 길리가 짝수라면 가운데 두글다를 출력하면 됩니다.

#입력값1 : "abcde" => c
#입력값3 : "qwer" => we

var = input("단어를 입력하라")
var_middle = int(len(var)/ 2 )
if len(var) % 2 != 0:
    print(var[var_middle])
else:
    print(var[var_middle-1:var_middle+1])
```
