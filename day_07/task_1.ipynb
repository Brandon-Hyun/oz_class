[문제 3]

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요.

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")


````python
user_list = []

while True:
    user_input = input("여러 개의 숫자를 입력하면 리스트에 저장됨 \n(공백으로 구분,'done'입력 시 종료)")

    # 종료 조건
    if user_input.lower() == "done":
        break

    # 입력값 처리
    try:
        # 공백으로 구분된 값들, 정수로 변환
        numbers = list(map(int, user_input.split()))
        user_list.extend(numbers)           # 리스트에 추가,append 는 리스트로 들어감 
    except ValueError:
        # 뭐가 문제냐
        for val in user_input.split():
            try:
                int(val)
            except ValueError:
                print(f'잘못된 값 : {val}. 숫자를 입력하세요')

print("최종 결과값 :", user_list)
````

[문제 6]

requests 모듈을 사용하여, 웹 페이지의 HTML을 가져오는 간단한 웹 스크레이퍼를 작성하세요.

'https://kream.co.kr/'에서<br>
데이터를 가져와서 상태 코드와 HTML의 첫 100자를 출력하세요.<br>
(실습 전 requests 모듈을 설치해야 함)

````python
import requests

url = "http://kream.co.kr/"

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) " "AppleWebKit/537.36 (KHTML, like Gecko) " "Chrome/120.0.0.0 Safari/537.36"}


try:
    response = requests.get(url, headers = headers, timeout = 10)


    print("status code:", response.status_code)

    # 요청이 성공시
    if response.ok:
        html_snippet = response.text[:100]
        print("html 의 첫 100 글자 :", html_snippet)
    else:
        print("요청 실패. 서버 응답:", response.reason)

except requests.exceptions.RequestException as exception:
    print("요청 중 오류 발생:", exception)
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요.

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def timer_decorator(func):
    def wrap(*args, **kwargs):
        time1 = time.time()
        result = func(*args, **kwargs)
        time2 = time.time() - time1
        print('함수 실행시간 : ',{time2},'초')
        return result
    return wrap

@timer_decorator
def count_some():
    for i in range(1000):
        print(i)

count_some()
````
