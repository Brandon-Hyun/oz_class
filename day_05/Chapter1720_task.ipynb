__[문제6]__

5번 문제에서 생성한 인스턴스의 이름, 나이, 성별을 출력해주세요.

```python
class Oz:
    def __init__(self, name:str, age:int, level:str):    #속성 초기화
        self.name = name     # 속성1 이름
        self.age = age      # 속성2 나이
        self.level = level      #속성3 레벨
    def coffee(self):
        print("뭐가 문젠데")        # 내가 모르는게 뭔지 모르겠음.

coding = Oz("김코딩",30,"A")

print(f'이름:{coding.name}\n나이:{coding.age}\n코딩레벨:{coding.level}')
```

__[문제7]__

오즈 (oz) 클래스에서 이름, 나이, 성별을 출력하는 who() 메소드를 추가하고 who()메소드를 이용해 아래 출력 결과를 만들어주세요


출력 결과 ) <br>
이름: 김코, 나이: 30, 코딩레벨: A

```python
class Oz:
    def __init__(self, name:str, age:int, level:str):    #속성 초기화
        self.name = name     # 속성1 이름
        self.age = age      # 속성2 나이
        self.level = level      #속성3 레벨
    def who(self):          #who 메서드 추가
        print(f'이름:{self.name}\n나이:{self.age}\n코딩레벨:{self.level}')        # 내가 모르는게 뭔지 모르겠음.

coding = Oz("김코딩",30,"A")
coding.who()
```
