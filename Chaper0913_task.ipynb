__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
nums_min = min(nums)
nums_max = max(nums)
nums_avg = sum(nums) / len(nums)
print(f'최소값은 {nums_min}입니다')
print(f'최대값은 {nums_max}입니다')
print(f'평균은 {nums_avg}입니다')

import statistics

average = statistics.mean(nums)
print(average)
````
