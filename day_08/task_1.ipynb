[문제 3]

'Book' 클래스를 선언하세요. 

이 클래스는 title (제목)과 author (저자) 두 개의 속성을 갖습니다. 

__str__ 특수 메소드를 오버라이드하여, 객체를 문자열로 변환할 때 "제목: [title], 저자: [author]" 형식의 문자열을 반환하도록 하세요. 

'Book' 객체를 하나 생성하고, print 함수를 사용하여 해당 객체의 문자열 표현을 출력하세요.

```python
class Book:
    def __init__(self,title, author):
        self.title = title
        self.author = author

    def __str__(self):                  # 문자열로 반환
        return f'제목 :[{self.title}], 저자 :[{self.author}]'
    
book01 = Book("K-Pop Demon Hunters", "Audrey 누나")         #인스턴스 book01 생성

print(str(book01))
```

[문제 4] 

'Employee' 클래스를 선언하세요. 

이 클래스는 월급인상률을 의미하는 클래스 변수로 'raise_amount'를 갖고 있으며, 기본값은 1.04입니다. 다음 급여는 기본 급여에 월급인상률을 곱한 값으로 결정합니다.

또한, 각 직원의 이름(name)과 급여(salary)를 속성으로 가지며, 'apply_raise' 메소드를 통해 급여에 'raise_amount'를 적용할 수 있어야 합니다.

'Employee' 클래스의 객체 두 개를 생성하고, 한 객체에만 'raise_amount'를 변경한 후, 두 객체의 급여를 증가시켜보세요.<br>
그리고 각 객체의 새로운 급여를 출력하세요.

```python
class Employee:
    raise_amount = 1.04                     # 클래스 변수 인상율

    def __init__(self, name, salary):       # 클래스 속성
        self.name = name
        self.salary = salary

    def apply_raise(self):      # 메서드
        self.salary *= self.raise_amount

    def get_salary(self):
        return self.salary

employee_01 = Employee("aaa", 5000)             # 객체 생성 1
employee_02 = Employee("bbb", 5000)             # 객체 생성 2

employee_01.raise_amount = 1.08                 # 인스턴스 변수로 오버라이드

employee_01.apply_raise()
employee_02.apply_raise()

print(employee_01.salary)
print(employee_02.salary)
```

[문제 5]

'Account' 클래스를 작성하세요. 

이 클래스는 소유자(owner)와 잔액(balance)을 속성으로 가지며, 잔액은 프라이빗 변수로 선언합니다. 

'deposit'와 'withdraw' 두 개의 메소드를 통해 잔액을 조작할 수 있도록 하세요. 

이 때, 잔액은 음수가 될 수 없도록 체크해야 합니다. 'Account' 객체를 생성하고,<br>
입금 및 출금 작업을 시뮬레이션한 후, 최종 잔액을 출력하세요.

```python
class Account:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance

    def deposit(self,amount):
        self.__balance += amount

    def withdraw(self,amount):
        if amount > self.__balance:
            print("잔액이 부족합니다. 현재 잔액은",self.__balance,"원 입니다.")
        else:
            self.__balance -= amount
    def get_balance(self):
        print(f'현재 잔액은 {self.__balance}원 입니다.')

client_01 = Account("김씨", 7000)
client_02 = Account("이씨", 8000)
client_03 = Account("학씨", 3000)

client_01.deposit(1000)
client_01.get_balance()
client_01.withdraw(9000)            # 입, 출금 후에도 잔액 나오게 출력해야겠다
client_01.get_balance()
client_01.withdraw(8000)
client_01.get_balance()
client_02.withdraw(6000)
client_02.get_balance()
client_02.deposit(-2000)
client_02.get_balance()
client_03.get_balance()                 # 누구의 계좌인지도?

```

[문제 6] 

'Vehicle'이라는 기본 클래스를 선언하세요. 

이 클래스는 make (제조사)와 model (모델명) 두 개의 속성을 갖고, 생성자(__init__)를 통해 이 두 속성을 초기화합니다. 'display_info'라는 메소드를 추가하여, "이 차량은 [make]에서 제조한 [model]입니다."라는 메시지를 반환하도록 하세요.

'Truck'과 'Car' 두 개의 하위 클래스를 생성하고, 각 클래스는 'Vehicle' 클래스로부터 상속받습니다. 

'Truck' 클래스에는 추가로 'payload' (적재량) 속성을, 'Car' 클래스에는 'seats' (좌석 수) 속성을 추가합니다. 각각의 하위 클래스에서 'display_info' 메소드는 기본 클래스의 메소드를 그대로 사용합니다.

'Truck'과 'Car' 객체 각각을 생성하고, 'display_info' 메소드를 호출하여 정보를 출력하세요. 

또한, 각 객체의 추가 속성(적재량, 좌석 수)에 대한 정보도 출력하세요.

이 문제는 클래스의 상속 개념을 이해하고, 기본 클래스의 속성과 메소드를 하위 클래스가 어떻게 활용할 수 있는지 보여줍니다. 

오버라이드는 필요하지 않으며, 각 하위 클래스의 고유한 속성을 추가하는 방법에 초점을 맞춥니다.

```python
class Vehicle:
    def __init__(self, maker, model):
        self.maker = maker
        self.model = model

    def display_info(self):
        print(f'이 차량은[{self.maker}]에서 제조한 [{self.model}]입니다.')

class Truck(Vehicle):
    def __init__(self, maker, model,payload):
        super().__init__(maker, model)
        self.payload = payload

    def display_info(self):
        super().display_info()
        print(f'적재량은 {self.payload}kg 입니다.')

class Car(Vehicle):
    def __init__(self, maker, model,seats):
        super().__init__(maker, model)
        self.seats = seats

    def display_info(self):
        super().display_info()
        print(f'좌석수 : {self.seats}인승 입니다.')

auto_001 = Vehicle("현대", "아반떼 CN7")
auto_002 = Vehicle("기아", "스포티지 NQ5")
auto_003 = Truck("기아", "봉고 프런티어", 2000)
auto_004 = Car("현대", "제네시스G80", 5)

auto_001.display_info()
auto_002.display_info()
auto_003.display_info()
auto_004.display_info()
```
