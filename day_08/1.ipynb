```python
class ChunkedRange:
    """"""

    """"""
    def __init__(self, stop, chunk_size):
        self.stop = stop
        self.chunked_size = chunk_size
        self.current = 0

    def __iter__ (self):
        return self

    def __next__(self):
        if self.current >= self.stop:
            raise StopIteration
        start = self.current
        end = min(self.current + self.chunked_size, self.stop)
        chunk = list(range(start, end))
        self.current = end
        return chunk
for chunk in ChunkedRange(stop=10, chunk_size=3):
    print(chunk)
```
