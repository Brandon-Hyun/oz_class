[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n):
    for i in range(n+1):
        if i % 2 ==1:
            yield i

for num in odd_numbers(20):
    print(num)
```

[문제 9] 주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(num):
    if num % 2 == 0:        # 조건 (짝수) 만족
        return True         # True 반환
    else:
        return False        # 조건 불만족 시 False 반환

print(is_even(7))
print(is_even(8))
print(is_even(9))
print(is_even(0))
print(is_even(0.7))          # 예외 적용해야되겠다.
print(is_even(-1))          #
print(is_even(-2))
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(num:int):
    if num == 0:                                    # 0! == 1
        return 1
    else:
        return num * factorial(num - 1)             #

num = int(input("숫자 입력"))                       # 입력

print(f'{num}! 는 {factorial(num)}입니다.')         # 결과 출력
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
# 구글링해서 어찌어찌 했는데 아직 소화 못함, 아니 재현도 못 할 듯..
def print_lengths(string:str):
    return [len(var_length) for var_length in string]                   # for 반복문, 리스트 string에서 순서대로 len(var)

var = ["의자", "책상", "책", "뭐긴단어", "더 긴 단어 뭐 없나 "]         # 입력될 리스트
print(print_lengths(var))
```
