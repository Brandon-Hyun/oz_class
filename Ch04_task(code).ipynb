__[문제 2: 첫 번째 강의 다리]__

<br>



돌문을 지나자 거친 강물이 흐르는 강이 나타났습니다.

아주 낡은 나무 다리가 힘겹게 메달려 있습니다.

당장이라도 무너질 것 같지만 이 다리 말고는 반대편으로 갈 수 있는 길이 보이지 않습니다.

무언가 불길한 기운이 계속 가까워지는 것만 같아 당신 빠르게 다리를 건너야 합니다.

다리는 50kg 이상 버티지 못하기 때문에 68kg인 당신은 무게를 줄여야 합니다.

방금 전 돌문을 통과하면서 얻은 마법 룬을 이용해 무게를 줄일 수 있습니다.

다리를 건너기 위해 줄여야 할 __최소 무게__ 를 구해주세요. (이상, 이하, 초과, 미만은 신경쓰지 않습니다!)


<br>

```python
# '내' 몸무게만 중요한지 아니면 다른 사람들의 몸무게도 중요해질 수 있는지?
# 아몰랑

weight_kg = 68
#몸무게, 마법으로 줄어든 무게, 마법이 줄여주는 무게 끼리 더하고 빼고 하면 순환참조 오류같은거 나나? 변수명은 적게 쓰면 좋은건가?

#마법이 줄여줘야하는 무게
magic_kg = 50-weight_kg
weight_kg = weight_kg + magic_kg

print(f' 마법으로 줄여야 할 최소 무게는 {abs(magic_kg)} kg 입니다.')
```

__[문제 4: 숲 속의 마법 연못]__

<br>



고목나무를 지나자 신비로운 마법 연못이 나타났습니다.

연못 속에는 불가사의한 에너지가 흐르고 있습니다.

이 에너지를 이용하면 보물을 찾는 데 필요한 단서를 얻을 수 있습니다.

연못의 깊이를 측정하려면 나무의 나이를 이용해야 합니다.

나무의 나이가 172살이고, 연못의 깊이는 나무 나이의 1/4입니다.

__연못의 깊이__ 를 구하세요.

<br>



```python
#나무의 나이는 172세인걸로
age_tree = pwd*4
#연못 깊이 = 나무나이 / 4
depth_pond = age_tree / 4
#길이의 단위를 알 수 없다. 웹브라우저에서 사용자의 국가정보나 사용언어 등을 읽어와서 미터법 안 쓰는 나라 필터 가능하겠네요.
print(f'연못의 깊이는 {depth_pond}이나 단위는 알 수 없습니다.')
```

__[문제 5: 사원의 입구]__

<br>



연못을 지나니 고대 사원이 나타났습니다. 사원의 입구에는 비밀번호를 입력해야 합니다.

비밀번호는 3자리 수로, 첫 번째 자리는 5, 두 번째 자리는 돌문을 여는 숫자의 십의 자리 수, 세 번째 자리는 돌문을 여는 숫자의 일의 자리 수의 3분의 1입니다.

__비밀번호__ 를 구하세요.

<br>

```python
# 십의자리 백의자리 숫자를 추출하는 방법?
# 세 번째 자리 수가 정수가 아니라면? 돌문을 여는 숫자의 일의 자리 수의 1/3이 1.1316 이면 어쩔,,,

# 10으로 나누면 몫이 십의 자리
tens = pwd // 10

# 일의 자리
units = ( pwd % 10 ) / 3

# print(f'비밀번호는 5{tens}{units}입니다.')
# 비밀번호는 541.0 이라고 뜨는데,자료형이 숫자이면 소수점 첫 째 자리까지 보여주는게 기본 설정인 듯
# 자료형을문자로 바꾸면 될 것 같은 느낌..

# 1의자리 + 십의자리 숫자*10 + 백의자리 숫자*100 하면 되나?
pwd_temple = int((5*100)+(10*tens)+units)
print(f'비밀번호는 {pwd_temple}입니다.')
```

__[문제 8: 폭포의 높이]__

<br>

섬의 반대편의 무사히 도착했습니다.

거대한 폭포가 당신의 길을 가로 막고 있습니다.

자세히 보니 폭포 위쪽에 반짝이는 무언가가 있습니다. 당신이 찾던 보물일 수 있습니다.

마법 룬을 이용해 폭포의 높이를 구하고 폭포의 위로 올라갑니다.

물방울이 떨어져 땅에 닿는 데 걸리는 시간은 2초입니다. 중력가속도는 9.8m/s²입니다.

돌문을 열때 사용한 숫자를 이용해 __폭포의 높이__ 를 구하세요. (단, 공기저항은 무시합니다.)

<br>


__Tips__
- 높이 = 0.5 * 중력가속도 * (시간 + 돌문의 숫자의 마지막 자리 수)²

<br>

```python
#길이 단위는 미터, 시간 단위는 초
height_meter = 0.5 * 9.8 * ((2 + pwd % 10 )**2)

print(f'폭포의 높이는 {height_meter} m 이다.')
```
